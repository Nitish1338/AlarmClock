```python
import tkinter as tk
from tkinter import messagebox

def start_countdown():
    try:
        hours = int(hours_entry.get())
        minutes = int(minutes_entry.get())
        seconds = int(seconds_entry.get())
        
        total_seconds = hours * 3600 + minutes * 60 + seconds
        countdown(total_seconds)
    except ValueError:
        pass

def countdown(seconds):
    if seconds >= 0:
        display_time(seconds)
        root.after(1000, countdown, seconds - 1)
    else:
        clear_entries()

def display_time(seconds):
    hours = seconds // 3600
    minutes = (seconds % 3600) // 60
    seconds = seconds % 60
    time_str.set(f"{hours:02d}:{minutes:02d}:{seconds:02d}")

def clear_entries():
    total_time = ""
    if int(hours_entry.get()) != 0:
        total_time += hours_entry.get() + " hour(s) "
    if int(minutes_entry.get()) != 0:
        total_time += minutes_entry.get() + " minute(s) "
    if int(seconds_entry.get()) != 0:
        total_time += seconds_entry.get() + " second(s) "
    
   
    hours_entry.delete(0, tk.END)
    minutes_entry.delete(0, tk.END)
    seconds_entry.delete(0, tk.END)
    hours_entry.focus_set()
    
    if total_time.strip() != "":
        messagebox.showinfo("Countdown Timer", f"Countdown Done for {total_time.strip()}")
    else:
        messagebox.showinfo("Countdown Timer", "Enter an appropriate value")

root = tk.Tk()
root.title("Countdown Clock")

time_str = tk.StringVar()
time_str.set("00:00:00") 
time_label = tk.Label(root, textvariable=time_str, font=("Helvetica", 48))
time_label.pack(padx=20, pady=20)

input_frame = tk.Frame(root)
input_frame.pack(padx=20, pady=10)

hours_label = tk.Label(input_frame, text="Hours:")
hours_label.grid(row=0, column=0)
hours_entry = tk.Entry(input_frame)
hours_entry.grid(row=0, column=1)
hours_entry.focus_set()

minutes_label = tk.Label(input_frame, text="Minutes:")
minutes_label.grid(row=1, column=0)
minutes_entry = tk.Entry(input_frame)
minutes_entry.grid(row=1, column=1)

seconds_label = tk.Label(input_frame, text="Seconds:")
seconds_label.grid(row=2, column=0)
seconds_entry = tk.Entry(input_frame)
seconds_entry.grid(row=2, column=1)

start_button = tk.Button(root, text="Start Timer", command=start_countdown)
start_button.pack(pady=10)

root.mainloop()
```
